```python
from transformers import BertTokenizer
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
tokens=tokenizer.tokenize('How tokenize tasks done?')
print(tokens)
```

```python
tokenizer.save_pretrained('my_tokenizer')
```

```python
tokenizer=BertTokenizer.from_pretrained('my_tokenizer')
tokens=tokenizer.tokenize('How tokenize tasks done?')
print(tokens)
input_ids=tokenizer.convert_tokens_to_ids(tokens)
print(input_ids)
final_prep=tokenizer.prepare_for_model(tokens)
print(final_prep['input_ids'])
inputs=tokenizer("How tokenize tasks done?")
print(tokenizer.decode(inputs['input_ids']))
```
