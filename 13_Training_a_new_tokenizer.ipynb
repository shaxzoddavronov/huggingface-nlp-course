```python
def get_training_corpus():
  return (
    data['train'][i:i+1000]['whole_func_string']
    for i in range(0,len(data['train']),1000) )
training_corpus=get_training_corpus()
```

```python
from transformers import AutoTokenizer
tokenizer=AutoTokenizer.from_pretrained("gpt2")
```

```python
print(data['train'][322]['whole_func_string'])
```

```python
sample='''
def message(self, msg):
        """Send a message to third party applications
        """
        for broker in self.message_brokers:
            try:
                broker(msg)
            except Exception as exc:
                utils.error(exc)'''
old_tokens=tokenizer.tokenize(sample)
print(old_tokens)
```

```python
new_tokenizer=tokenizer.train_new_from_iterator(training_corpus,52000)
```

```python
tokens=new_tokenizer.tokenize(sample)
print(tokens)
```
