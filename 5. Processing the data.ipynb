```python
from datasets import load_dataset
from transformers import AutoTokenizer
```

```python
data=load_dataset('glue','mrpc')
data
```

```python
checkpoint='bert-base-uncased'
tokenizer=AutoTokenizer.from_pretrained(checkpoint)

def tokenizer_function(sample):
  return tokenizer(sample['sentence1'],sample['sentence2'],padding='max_length',truncation=True,max_length=128)

raw_tokenized=data.map(tokenizer_function,batched=True)
raw_tokenized
```

```python
raw_tokenized=raw_tokenized.remove_columns(['sentence1','sentence2','idx'])
raw_tokenized=raw_tokenized.rename_column('label','labels')
raw_tokenized=raw_tokenized.with_format('torch')
raw_tokenized['train'][0]
```

```python
from torch.utils.data import DataLoader
from transformers import DataCollatorWithPadding
```

```python
dataloader_train=DataLoader(raw_tokenized['train'],batch_size=24,shuffle=True)
for step,batch in enumerate(dataloader_train):
  print(batch['input_ids'].shape)
  if step>3:
    break
```

```python
checkpoint='bert-base-uncased'
tokenizer=AutoTokenizer.from_pretrained(checkpoint)

def tokenizer_function(sample):
  return tokenizer(sample['sentence1'],sample['sentence2'],truncation=True)

raw_tokenized=data.map(tokenizer_function,batched=True)
raw_tokenized=raw_tokenized.remove_columns(['sentence1','sentence2','idx'])
raw_tokenized=raw_tokenized.rename_column('label','labels')
raw_tokenized=raw_tokenized.with_format('torch')

data_collator=DataCollatorWithPadding(tokenizer=tokenizer)
dataloader_train=DataLoader(raw_tokenized['train'],batch_size=24,shuffle=True,collate_fn=data_collator)
for step,batch in enumerate(dataloader_train):
  print(batch['input_ids'].shape)
  if step>3:
    break
```
