```python
from datasets import load_dataset
file_data={"train":"/content/drugsComTrain_raw.tsv","test":"/content/drugsComTest_raw.tsv"}
drugs_dataset=load_dataset("csv",data_files=file_data,delimiter='\t')
drugs_dataset
```

```python
sample_drug=drugs_dataset['train'].shuffle(seed=23).select(range(2000))
sample_drug
```

```python
sample_drug=sample_drug.rename_column('Unnamed: 0','patient_id')
sample_drug
```

```python
sample_drug[0]
```

```python
def review_length_map(sample):
  return {'review_length':len(sample['review'].split())}
sample_drug=sample_drug.map(review_length_map)
sample_drug
```

```python
#sample_drug=sample_drug.train_test_split(test_size=0.2)
def split_tokenize(sample):
  return tokenizer(sample['review'],
                   truncation=True,
                   max_length=128,
                   return_overflowing_tokens=True)
tokenized_sample=split_tokenize(sample_drug['train'][4])
x=[len(ids) for ids in tokenized_sample['input_ids']]
x
```

```python
tokenized_drug=sample_drug.map(split_tokenize,batched=True)
tokenized_drug
```

```python
tokenized_drug=sample_drug.map(split_tokenize,batched=True,remove_columns=sample_drug['train'].column_names)
tokenized_drug
```

```python
def tokenized_split(sample):
  result=tokenizer(sample['review'],truncation=True,
                   max_length=128,return_overflowing_tokens=True)

  sample_map=result.pop('overflow_to_sample_mapping')
  for key,value in result.items():
    result[key]=[value[i] for i in sample_map]
  return result

tokenized_drug=sample_drug.map(tokenized_split,batched=True)
tokenized_drug
```
